```python
x=[1,6,1,6,8,9,0,4,6,8,9,4,3,6,8,0,12,8,23,7,6,34,45,3,34,5,56,3,2,5,4,32,42,53,35,24,65,7,89,37,35,5,17,28,59,6,85,23]

def qsort(myx):
    x=[]
    y=[]
    z=[]
    for i in myx:
        if i<myx[0]: x.append(i)
        elif i==myx[0]: y.append(i)
        else: z.append(i)
    if len(x)>1: x=qsort(x)
    if len(z)>1: z=qsort(z)
    return x+y+z

print qsort(x)

quicksort = lambda seq: [] if not seq else quicksort(filter(lambda n: n<=seq[0], seq[1:]))+[seq[0]]+quicksort(filter(lambda n: n>seq[0], seq[1:]))
print quicksort(x)
```

```python
def myfunc (st, en, x=50, endpoint=True, retstep=False):
    z=[]
    z.append(st)
    x1=x-1
    if endpoint: x1=x
    up=(en-st)/(float(x1))
    for i in range(1,x): z.append(z[i-1]+up)
    if retstep: z=(z,up)
    return z
    
print myfunc(2,5,3)
print myfunc(2,5,3,False)
print myfunc(2,5,3,retstep=True)

V = lambda st, en, x=50, endpoint=True, retstep=False: ([st+i*(en-st)/float(x-1+int(endpoint)) for i in range(0,x)],int(retstep)*(en-st)/float(x-1+int(endpoint)))

print V(2,5,3)
print V(2,5,3,False)
print V(2,5,3,retstep=True)

from numpy import *

%timeit a = array( [2,3,4] )

%timeit a = zeros( (3,4) )
a = zeros( (3,4) )
print a

a = array( [[20,30],[40,50]] )
b = array( [[1,2],[3,4]] )
c = a-b
%timeit c=a-b
print a
print b
print c

a = array([[1,0],[0,1]])
b = array([[2,3],[4,5]])
c = a*b
print a
print b
print c
%timeit c=a*b

a = array([[0,1],[0,1]])
b = array([[2,3],[4,5]])
c = dot(a,b)
print a
print b
print c
print dot(a,b)
%timeit c = dot(a,b)

%timeit a = random.random((2,3))
a = random.random((2,3))
%timeit a.sum()
a.sum()
print a
print a.sum()
print a.min(axis=1)
print a.min(axis=0)
```
