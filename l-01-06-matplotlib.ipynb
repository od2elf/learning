```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)
                   
z=[np.sin(i/2.) for i in range(0,51)]
y=[i for i in range(0,51)]
x=[i*i/100. for i in range(0,51)]

ax1.plot(x, label='pow2')
ax1.annotate('pow2', xy=(25,x[25]), xytext=(25-20, x[25]+5),arrowprops=dict(facecolor='black', shrink=0.01),)
ax1.plot(y, label='linear')
ax1.annotate('linear', xy=(25,y[25]), xytext=(25-20, y[25]+5),arrowprops=dict(facecolor='black', shrink=0.05),)
ax1.plot(z, label='sin')
ax1.annotate('sin', xy=(25,z[25]), xytext=(25-20, z[25]-5),arrowprops=dict(facecolor='black', shrink=0.05),)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

ax2 = fig.add_subplot(1,2,2)

ax2.plot(x, label='pow2')
ax2.annotate('pow2', xy=(25,x[25]), xytext=(25-20, x[25]+5),arrowprops=dict(facecolor='black', shrink=0.01),)
ax2.plot(y, label='linear')
ax2.annotate('linear', xy=(25,y[25]), xytext=(25-20, y[25]+5),arrowprops=dict(facecolor='black', shrink=0.05),)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()

plt.show()

fig.savefig('pic.png', format='png')
```
